# Библиотека H20

**H2O** - это открытая платформа для анализа данных и машинного обучения, которая обеспечивает широкий спектр алгоритмов машинного обучения, включая глубокое обучение, градиентный бустинг, случайный лес и другие.

AutoML (автоматическое машинное обучение) - это подход к машинному обучению, который автоматизирует процесс построения модели машинного обучения. AutoML в H2O работает так:

1. **Загрузка данных**: В первую очередь вам нужно загрузить свой набор данных в H2O.

2. **Запуск AutoML**: Далее вы можете запустить процесс AutoML, указав, какие столбцы данных использовать в качестве функций, а какой - в качестве целевого. Вы также можете указать, сколько времени вы хотите, чтобы AutoML работал.

3. **Подбор моделей**: AutoML автоматически обучает и валидирует множество моделей, используя ваш набор данных. Он использует различные алгоритмы и методы для построения и оптимизации моделей, включая предобработку данных, отбор функций, настройку гиперпараметров и стекинг моделей.

4. **Ранжирование моделей**: По завершении процесса AutoML модели ранжируются в соответствии с их производительностью на основе выбранной метрики (например, точности или ошибки MSE).

5. **Использование модели**: Вы можете выбрать лучшую модель для дальнейшего использования, а также получить ее параметры и информацию о производительности.

Одно из ключевых преимуществ использования AutoML - это то, что он может облегчить процесс построения модели машинного обучения для тех, кто не имеет глубоких знаний в области машинного обучения. Он также может помочь экспертам по машинному обучению быстро создать базовые модели и ускорить процесс исследования и разработки.

## Пример №1 -  датасет симуляционных данных, которые описывают разложение бозона Хиггса

Работу библиотеки H2O посмотрим на примере датасета симуляционных данных, которые описывают разложение бозона Хиггса. Данный датасет связан с задачей классификации в области физики высоких энергий и был использован в конкурсе Higgs Boson Machine Learning Challenge на платформе Kaggle в 2014 году.

Он представляет собой набор симуляционных данных, которые описывают разложение бозона Хиггса и других процессов в частицевом детекторе. Бозон Хиггса - это частица, предсказанная теорией стандартной модели частиц, и его открытие было одним из ключевых моментов в физике частиц.

В датасете содержатся различные физические параметры этих процессов, включая информацию о моменте импульса частиц, углах между частицами и других характеристик.

Каждый образец в датасете представляет собой событие, которое может быть либо сигналом (т.е. разложением бозона Хиггса), либо фоном (т.е. другим процессом). Задача состоит в том, чтобы построить модель, которая может различить эти два класса на основе доступных данных.

```python
# Import a sample binary outcome train/test set into H2O
train = h2o.import_file("https://s3.amazonaws.com/erin-data/higgs/higgs_train_10k.csv")
test = h2o.import_file("https://s3.amazonaws.com/erin-data/higgs/higgs_test_5k.csv")
```

```python
train
```

```python
# Определение предикторов и целевой переменной
x = train.columns  # Получаем список всех колонок в данных обучения
y = "response"  # Указываем имя целевой переменной
x.remove(y)  # Удаляем из списка предикторов имя целевой переменной

# Для бинарной классификации целевая переменная должна быть фактором (категорией)
train[y] = train[y].asfactor()  # Преобразуем целевую переменную в данные обучения в фактор
test[y] = test[y].asfactor()  # Преобразуем целевую переменную в тестовых данных в фактор
```

В данном коде мы готовим данные для обучения модели машинного обучения. Вначале мы определяем список предикторов и целевую переменную. Предикторы - это переменные, которые мы используем для прогнозирования целевой переменной.

После этого мы преобразуем целевую переменную response в фактор (или категорию) в обучающем и тестовом наборах данных. Это делается, потому что наша задача - бинарная классификация, и модель должна предсказывать категорию, а не непрерывное значение.

В H2O AutoML метрика, которую нужно оптимизировать, выбирается автоматически на основе типа задачи.

Если задача является задачей классификации, как в нашем случае, H2O AutoML будет автоматически оптимизировать метрику logloss.
Для задач регрессии используется deviance.

Важно отметить, что H2O AutoML всегда стремится максимизировать метрику, поэтому для метрик, где меньшие значения лучше (например, logloss или MSE), H2O использует их отрицательные значения.

Если вы хотите использовать другую метрику для оптимизации, вы можете задать параметр stopping_metric при создании экземпляра H2OAutoML.

```python
# Запуск AutoML для 20 базовых моделей
aml = H2OAutoML(max_models=20, seed=1)  # Создаем объект AutoML, устанавливаем максимальное количество моделей в 20 и задаем начальное значение для генератора случайных чисел

aml.train(x=x, y=y, training_frame=train)  # Обучаем модели на заданном обучающем датасете с указанными предикторами (x) и целевой переменной (y)
```

После выполнения этого кода мы получаем обученный объект AutoML (в нашем случае, он называется "aml"), который содержит до 20 обученных моделей, основанных на различных алгоритмах машинного обучения. Эти модели обучены на нашем обучающем наборе данных (в данном случае, train) для предсказания целевой переменной (y), используя все остальные столбцы в данных в качестве предикторов (x).

"aml" также содержит "доску лидеров" - таблицу, которая ранжирует все модели, обученные в процессе AutoML, на основе выбранной метрики (по умолчанию AUC для бинарной классификации, mean_per_class_error для многоклассовой классификации и deviance для регрессии).

Самая сильная модель (то есть модель с наилучшим показателем по выбранной метрике) становится "лидером" и может быть использована для предсказания новых данных. Вы можете получить эту модель, используя aml.leader.

Длительность поиска H2O AutoML определяется либо согласно указанному  времени работы алгоритма (параметр max_runtime_secs), либо на основе количества моделей (параметр max_models). Если установлены оба параметра, выполнение AutoML остановится, как только будет достигнуто одно из этих значений. 

### Необязательные параметры:

* **x**: Список или вектор имен или индексов предикторных колонок. Этот аргумент нужно указывать только в случае, если вы хотите исключить некоторые колонки из набора предикторов. Если все колонки (кроме целевой переменной) должны использоваться для прогнозирования, этот аргумент не нужно указывать.

* **nfolds**: Задает количество фолдов (разбиений) для кросс-валидации моделей в AutoML. Значение должно быть >= 2 или "-1", чтобы AutoML самостоятельно выбрал между кросс-валидацией и смешиванием (blending). Смешивание будет использовать часть training_frame (если не предоставлен blending_frame) для обучения стековых ансамблей. Установите 0, чтобы отключить кросс-валидацию; это также отключит стековые ансамбли (что может снизить общую эффективность модели). По умолчанию установлено значение "-1".

* **balance_classes**: Указывает, следует ли увеличивать количество примеров миноритарных классов для балансировки распределения классов. Этот параметр не включен по умолчанию и может увеличить размер набора данных. Этот параметр применим только для классификации. Если размер датасета после оверсемплинга превышает максимальный размер, рассчитанный с помощью параметра max_after_balance_size, то большинство классов будет досэмплировано, чтобы удовлетворить ограничению по размеру.

* **class_sampling_factors**: Указывает отношения переоценки / недооценки для каждого класса (в лексикографическом порядке). По умолчанию эти отношения автоматически рассчитываются во время обучения для достижения баланса классов. Обратите внимание, что для этого требуется установить balance_classes в True.

* **max_after_balance_size**: Задает максимальный относительный размер обучающих данных после балансировки классов (balance_classes должен быть включен). По умолчанию равно 5.0. (Значение может быть меньше 1.0).

* **max_runtime_secs_per_model**: Указывает максимальное количество времени, выделенное на обучение каждой отдельной модели в AutoML. По умолчанию равно 0 (отключено). Обратите внимание, что модели с ограничением по времени не гарантируют воспроизводимость.

* **stopping_metric**: Задает метрику для ранней остановки. По умолчанию равно AUTO. Доступные опции:

*   *AUTO*: По умолчанию выбирается *logloss* для классификации и *deviance* для регрессии.
*   *deviance* (среднее остаточное отклонение)
*   *logloss*
*   *MSE*
*   *RMSE*
*   *MAE*
*   *RMSLE*
*   *AUC (площадь под ROC-кривой)*
*   *AUCPR* (площадь под кривой точность-полнота)
*   *lift_top_group*
*   *misclassification*
*   *mean_per_class_error*

... и другие параметры.

Эти параметры обеспечивают гибкость при настройке процесса AutoML для вашего конкретного набора данных и задачи.

```python
# Можно посмотреть на результаты других моделей - AutoML Leaderboard
lb = aml.leaderboard
lb.head(rows=lb.nrows)  # Выведем все строки
```

где /your/destination/path - это путь к директории, где вы хотите сохранить модель. Если директории не существует, она будет создана.

```python
# Можно получить ключи, которые относятся к параметрам лучшей модели, которую мы получили с помощью H2O AutoML:
best_model.params.keys()
```

Значения этих параметров можно получить, обратившись к каждому из них через объект модели. В данном случае, мы получаем только ключи словаря параметров, но чтобы использовать эту информацию, нам нужно также получить и соответствующие значения.

```python
metalearner_algorithm_value = best_model.params['metalearner_algorithm']['actual']
print(metalearner_algorithm_value)
```

Здесь 'actual' используется для извлечения фактического значения параметра, потому что best_model.params['metalearner_algorithm'] вернет словарь, содержащий дополнительную информацию о параметре, включая его 'default' значение и 'actual' значение.

В общем случае, вы можете пройтись по всем ключам и получить их значения:

```python
for key in best_model.params.keys():
    print(key, best_model.params[key]['actual'])
```

## Пример №2 - датасет на тему прогнозирования задержек в доставке товаров (product backorders)

Рассмотрим работу H2OAutoML также на примере набора данных для прогнозирования возможного дефицита продукции (backorders). Вот краткое описание столбцов в этом наборе данных:

* **sku** - это SKU (Stock Keeping Unit), уникальный идентификатор товара для учета на складе.

* **national_inv** - "национальный запас", количество данного товара, доступного на национальном уровне.

* **lead_time** - это время от начала процесса (например, заказа товара у поставщика) до конца (например, получения товара).

* **in_transit_qty** - это количество товара, которое в данный момент находится в пути (в процессе перевозки).

* **forecast_3_month, forecast_6_month, forecast_9_month** - это прогнозы потребности в товаре на следующие 3, 6 и 9 месяцев.

* **sales_1_month, sales_3_month, sales_6_month, sales_9_month** - это данные о продажах данного товара за последний месяц, 3 месяца, 6 месяцев и 9 месяцев соответственно.

* **min_bank** - минимальный запас товара (минимальное количество, которое должно быть на складе).

* **potential_issue, pieces_past_due, perf_6_month_avg, perf_12_month_avg, local_bo_qty** - информация, которая относится к возможным проблемам с поставками, количеству товаров, поставка которых задерживается, среднему показателю выполнения заказов за последние 6 и 12 месяцев и количеству товаров, которые не доставлены на локальном уровне.

* **went_on_backorder** - это целевая переменная, которая указывает, ушел ли данный товар в дефицит (да или нет).

```python
import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

### Загружаем датасет

```python
# Use local data file or download from GitHub
import os
docker_data_path = "/home/h2o/data/automl/product_backorders.csv"
if os.path.isfile(docker_data_path):
  data_path = docker_data_path
else:
  data_path = "https://github.com/h2oai/h2o-tutorials/raw/master/h2o-world-2017/automl/data/product_backorders.csv"


# Load data into H2O
df = h2o.import_file(data_path)
```

### Готовим выборки

```python
df
```

Теперь давайте определим столбцы с правильными ответами и предикторов, сохранив их как x и y. Столбец «sku» — это уникальный идентификатор, поэтому мы  удаляем его из набора наших предикторов.

```python
y = "went_on_backorder"
x = df.columns
x.remove(y)
x.remove("sku")
```

Запустим AutoML, указав, что будет запущено 10 моделей. Аргумент max_models указывает количество отдельных (или «базовых») моделей и не включает две модели ансамбля, которые обучаются в конце.

Для каждой обученной модели вычисляются метрики качества с использованием кросс-валидации, и затем модели ранжируются на основе этих метрик. Например, модель с наивысшим средним AUC-ROC на кросс-валидации будет на вершине списка. Этот подход позволяет выбирать модели, которые наиболее вероятно будут хорошо работать на новых, неизвестных данных.

```python
aml_2 = H2OAutoML(max_models = 10, seed = 1)
aml_2.train(x = x, y = y, training_frame = df)
```

```python
# получаем лучшую модель:
best_model_2 = aml_2.leader
```

Выведем информацию по всем моделям. Здесь мы увидим два ансамбля моделей в верхней части таблицы лидеров. ансамбли почти всегда могут превзойти одну модель.

**Stacked Ensembles** - это метод моделирования, который объединяет различные модели машинного обучения для создания одной, более мощной модели. В основе стекинга лежит идея, что несколько слабых моделей, работающих вместе, могут быть более точными и устойчивыми, чем одна отдельная модель.

```python
lb = aml_2.leaderboard
lb
```

```python
os.listdir(IMAGE_PATH)
```

```python
# Создаем генератор данных для нормирования и аугментации тренировочной выборки
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,      # диапазон углов для вращения изображений
    width_shift_range=0.2,  # диапазон, который используется для горизонтального смещения изображения
    height_shift_range=0.2, # диапазон, который используется для вертикального смещения изображения
    zoom_range=0.2,         # диапазон для случайного масштабирования изображения
    fill_mode='nearest',     # стратегия заполнения появившихся пустых мест при преобразованиях
    validation_split=0.2)
```

## Пример №1: библиотека Optuna для оптимизации гиперпараметров с использованием тюнера TPESampler

```python
def create_model(trial):
    # Определение гиперпараметров, которые будут оптимизироваться
    learning_rate = trial.suggest_float('learning_rate', 1e-5, 1e-1, log=True)
    num_filters = trial.suggest_int('num_filters', 16, 128)
    dropout_rate = trial.suggest_float('dropout_rate', 0.0, 0.5)
    dense_units = trial.suggest_categorical('dense_units', [64, 128, 256])
    activation = trial.suggest_categorical('activation', ['relu', 'tanh'])
    use_batchnorm = trial.suggest_categorical('use_batchnorm', [True, False])
    use_dropout = trial.suggest_categorical('use_dropout', [True, False])
    
    # Создание модели нейросети
    model = keras.Sequential()
    model.add(layers.Conv2D(num_filters, 3, activation=activation, input_shape=(64, 64, 3)))
    model.add(layers.MaxPooling2D())
    if use_batchnorm:
        model.add(layers.BatchNormalization())
    if use_dropout:
        model.add(layers.Dropout(dropout_rate))
    model.add(layers.Conv2D(num_filters, 3, activation=activation))
    model.add(layers.MaxPooling2D())
    if use_batchnorm:
        model.add(layers.BatchNormalization())
    if use_dropout:
        model.add(layers.Dropout(dropout_rate))
    model.add(layers.Flatten())
    model.add(layers.Dense(dense_units, activation=activation))
    model.add(layers.Dropout(dropout_rate))
    model.add(layers.Dense(3, activation='softmax'))

    # Компиляция модели
    model.compile(optimizer=keras.optimizers.Adam(learning_rate),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    return model
```

```python
def objective(trial):
    # Создание модели с определенными гиперпараметрами
    model = create_model(trial)

    # Обучение модели
    model.fit(train_data, validation_data=val_data, epochs=10)

    # Оценка модели на валидационной выборке
    loss, accuracy = model.evaluate(val_data)

    return accuracy
```

В библиотеке Optuna можно выбрать определенный тюнер при создании объекта Study с помощью параметра sampler. Тюнер определяет, каким образом будет исследоваться пространство поиска гиперпараметров.

Доступные тюнеры в Optuna:

* **TPESampler**: Тюнер, основанный на Tree-structured Parzen Estimator (TPE). 
* **RandomSampler**: Тюнер, выбирающий значения гиперпараметров случайным образом.
* **CmaEsSampler**: Тюнер, использующий алгоритм Covariance Matrix Adaptation Evolution Strategy (CMA-ES) для оптимизации.
* **и другие тюнеры (сэмплеры)** - подробнее рассмотрим на занятии




```python
sampler = optuna.samplers.TPESampler()
# Создание study объекта для оптимизации
study = optuna.create_study(direction='maximize', sampler=sampler)

# Запуск оптимизации гиперпараметров
study.optimize(objective, n_trials=8)
```

## Пример №2: библиотека Optuna для оптимизации гиперпараметров с использованием прунеров (стратегии обрезки)

Модуль **"pruners"** в библиотеке Optuna предоставляет классы для стратегий обрезки (pruning) в процессе оптимизации. Обрезка - это техника, которая позволяет прекратить дальнейшее выполнение оптимизации для определенных пробных точек, которые, вероятно, не будут приводить к улучшению результатов. Процесс обрезки позволяет экономить вычислительные ресурсы, ускоряет оптимизацию и улучшает сходимость.

Базовый класс "BasePruner" определяет абстрактный метод "prune()", который принимает объект "trial" и связанное с ним исследование "study" и возвращает логическое значение, указывающее, следует ли обрезать (прекратить выполнение) эту пробную точку. Определение обрезки основано на промежуточных значениях целевой функции, предварительно сообщенных для пробы с использованием метода "optuna.trial.Trial.report()".

В модуле "pruners" также содержатся дочерние классы, которые наследуются от "BasePruner" и реализуют различные стратегии обрезки, такие как "MedianPruner", "PercentilePruner" и "SuccessiveHalvingPruner". Каждый из этих классов имеет свои уникальные алгоритмы и параметры, определяющие условия обрезки пробных точек.

Для использования Pruner в библиотеке Optuna, мы должны указать его в качестве аргумента при создании объекта Study.

Посмотрим на пример использования MedianPruner, который прерывает неуспешные пробные запуски, если их значение функции превышает медианное значение. Однако в библиотеке Optuna также доступны другие прунеры, такие как SuccessiveHalvingPruner, HyperbandPruner и другие, которые могут быть использованы вместо MedianPruner в зависимости от ваших потребностей.

Параметр **n_startup_trials** означает, что выбранный прунер начнет работать, когда завершится заданное количество испытаний в одном и том же исследовании.

```python
# Создание прунера
pruner = optuna.pruners.MedianPruner(n_startup_trials=5)

# Определение модели и гиперпараметров
def objective(trial):
    # Определение параметров, которые будут оптимизироваться
    # Определение гиперпараметров, которые будут оптимизироваться
    learning_rate = trial.suggest_float('learning_rate', 1e-5, 1e-1, log=True)
    num_filters = trial.suggest_int('num_filters', 8, 256)
    dropout_rate = trial.suggest_float('dropout_rate', 0.0, 0.5)
    dense_units = trial.suggest_categorical('dense_units', [16, 256])
    activation = trial.suggest_categorical('activation', ['relu', 'tanh'])
    use_batchnorm = trial.suggest_categorical('use_batchnorm', [True, False])
    use_dropout = trial.suggest_categorical('use_dropout', [True, False])
    
    # Создание модели нейросети
    model = keras.Sequential()
    model.add(layers.Conv2D(num_filters, 3, activation=activation, input_shape=(64, 64, 3)))
    model.add(layers.MaxPooling2D())
    if use_batchnorm:
        model.add(layers.BatchNormalization())
    if use_dropout:
        model.add(layers.Dropout(dropout_rate))
    model.add(layers.Conv2D(num_filters, 3, activation=activation))
    model.add(layers.MaxPooling2D())
    if use_batchnorm:
        model.add(layers.BatchNormalization())
    if use_dropout:
        model.add(layers.Dropout(dropout_rate))
    model.add(layers.Flatten())
    model.add(layers.Dense(dense_units, activation=activation))
    model.add(layers.Dropout(dropout_rate))
    model.add(layers.Dense(3, activation='softmax'))

    # Компиляция модели
    model.compile(optimizer=keras.optimizers.Adam(learning_rate),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    # на скольки эпохах будет запускаться каждая модель
    num_epochs = 10
    for epoch in range(num_epochs):
        # Обучение модели
        history  =model.fit(train_data, validation_data=val_data, batch_size=32, epochs=1, verbose=1)
        # Получение значений функции потерь и метрик на каждой эпохе
        val_accuracy = history.history['accuracy'][-1]
        trial.report(val_accuracy, step=epoch)
        if trial.should_prune():
            raise optuna.TrialPruned()
    return val_accuracy 
```

Теперь, если обучение модели будет идти плохо (в смысле точности), trial.should_prune() вернёт True, и будет вызвано исключение optuna.TrialPruned(), что приведёт к прерыванию обучения текущей модели и переходу к следующему испытанию.

```python
# Создание study с указанием прунера
study = optuna.create_study(direction='maximize', pruner=pruner)
# Запуск оптимизации
study.optimize(objective, n_trials=10)
```

В этом коде каждый испытательный набор гиперпараметров (модель) обучается на протяжении 10 эпох, и после каждой эпохи проверяется условие прунинга. Если условие прунинга выполняется, текущее испытание обрывается, и Optuna переходит к следующему набору гиперпараметров.

```python
# Получение лучшего набора гиперпараметров
best_params = study.best_params
best_accuracy = study.best_value
print('Лучшие гиперпараметры:', best_params)
print('Лучшая точность:', best_accuracy)
```

## Сравнение "Early stopping" и "pruning"

Callback EarlyStopping в Keras и Pruners в Optuna оба используются для преждевременного прекращения обучения, но они имеют некоторые ключевые отличия:

* **Callback EarlyStopping** в Keras: Этот callback прекращает обучение, когда метрика не улучшается в течение заданного числа эпох. Когда происходит ранняя остановка, последняя и лучшая модель может быть автоматически сохранена. Это полезно, когда вы обучаете одну большую модель и хотите остановить обучение, как только оно перестает улучшаться.

* **Pruners** в Optuna: Pruners применяются во время оптимизации гиперпараметров, чтобы прекратить выполнение определенных испытаний, которые вряд ли приведут к лучшим результатам. Они используются для увеличения эффективности оптимизации гиперпараметров, позволяя более быстро исследовать пространство гиперпараметров.

Сходства:
Оба инструмента используются для оптимизации процесса обучения, прекращая обучение, когда оно, вероятно, больше не приведет к улучшению.

Различия:

Callback EarlyStopping применяется для отдельной модели, в то время как Pruners в Optuna используются при оптимизации гиперпараметров для большого числа моделей.
EarlyStopping применяется на уровне эпох при обучении одной модели, в то время как Pruners в Optuna применяются на уровне испытаний при оптимизации гиперпараметров.
EarlyStopping сохраняет лучшую модель, в то время как Pruners просто прекращают испытания.
Optuna Pruners могут использовать различные стратегии для определения того, должно ли быть прервано испытание, включая отслеживание прогресса по сравнению с другими испытаниями (Median Pruner), использование информации о предыдущих испытаниях (SuccessiveHalving Pruner), и др.

**Краткое описания прунеров Optuna:**

* **BasePruner**: Базовый класс для прунеров. Все другие прунеры наследуют от него свои свойства и методы.

* **MedianPruner**: Применяет правило медианной остановки. Если результат промежуточной метрики испытания ниже медианы из определенного количества предыдущих испытаний, то это испытание прерывается.

* **NopPruner**: Прунер, который никогда не обрезает испытания. Полезен для отключения обрезки без изменения кода обучения.

* **PatientPruner**: Прунер, который оборачивает другой прунер, добавляя терпимость к его решениям об обрезке. В случае, если другой прунер решает обрезать испытание, PatientPruner будет ждать заданное количество шагов прежде, чем это испытание фактически будет обрезано.

* **PercentilePruner**: Прунер, который сохраняет указанный процентил испытаний. Если результат промежуточной метрики испытания ниже заданного процентиля среди всех промежуточных результатов этого шага в предыдущих испытаниях, то это испытание прерывается.

* **SuccessiveHalvingPruner**: Применяет алгоритм Асинхронного Последовательного Уменьшения. Этот прунер является многозадачным и обрезает каждое испытание на определенных этапах, основываясь на его промежуточных результатах.

* **HyperbandPruner**: Применяет алгоритм Hyperband. Этот прунер также многозадачен и использует SuccessiveHalvingPruner на разных уровнях ресурсов.

* **ThresholdPruner**: Прунер, который обнаруживает отклоняющиеся метрики испытаний. Если промежуточная метрика испытания превышает или уступает заданному порогу, то это испытание прерывается.

# Библиотека Hyperopt

Мануал https://github.com/hyperopt/hyperopt/wiki/FMin

Hyperopt — это библиотека Python для оптимизации гиперпараметров, которая использует поиск по дереву для быстрого и эффективного поиска в пространстве гиперпараметров. В Hyperopt существуют три основных компонента:

Objective function: Это функция, которую вы хотите оптимизировать. Обычно она представляет собой функцию потерь, которую нужно минимизировать.

Domain space: Это пространство возможных значений гиперпараметров. Вы можете определить разные типы пространств, такие как дискретные, непрерывные или категориальные.

Optimization algorithm: Это алгоритм, который будет искать в пространстве гиперпараметров. В Hyperopt доступны разные алгоритмы, такие как случайный поиск и дерево Парзеновских оценщиков (Tree of Parzen Estimators, TPE).

Вот пример использования Hyperopt для многоклассовой классификации изображений с помощью нейронной сети Keras:

```python
from hyperopt import fmin, tpe, hp, STATUS_OK, Trials
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, BatchNormalization, MaxPooling2D, Conv2D
from keras.optimizers import Adam
```

```python
# Определение целевой функции для оптимизации
def objective(params):
    model = Sequential()
    # Входной сверточный слой
    model.add(Conv2D(filters=params['filters'], kernel_size=params['kernel_size'], 
                     activation=params['activation'], input_shape=(64, 64, 3)))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=params['pool_size']))
    model.add(Dropout(params['dropout']))

    # Второй сверточный слой
    model.add(Conv2D(filters=params['filters']*2, kernel_size=params['kernel_size'], 
                     activation=params['activation']))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=params['pool_size']))
    model.add(Dropout(params['dropout']))

    model.add(Flatten())
    
    # Полносвязный слой
    model.add(Dense(params['units'], activation=params['activation']))
    model.add(Dropout(params['dropout']))

    model.add(Dense(3, activation='softmax'))

    model.compile(loss='categorical_crossentropy', 
                  optimizer=Adam(lr=params['learning_rate']),
                  metrics=['accuracy'])

    history = model.fit(train_data, epochs=10, batch_size=params['batch_size'], 
                        validation_data=val_data, verbose=0)
    
    # Получаем точность валидации на последней эпохе
    val_acc = history.history['val_accuracy'][-1]
    return {'loss': -val_acc, 'status': STATUS_OK, 'Trained_Model': model} 
```

Пространство гиперпараметров определяется при помощи следующих инструментов:

* **hp.choice**(label, options): Возвращает один из предложенных вариантов из списка или кортежа. Элементы options могут быть стохастическими выражениями.

* **hp.randint**(label, upper): Возвращает случайное целое число в диапазоне (0, upper). Подходит для описания случайных значений, например, seed-ов.

* **hp.uniform**(label, low, high): Возвращает значение, равномерно распределенное между low и high.

* **hp.quniform**(label, low, high, q): Возвращает значение вроде round(uniform(low, high) / q) * q. Подходит для дискретных значений, которые должны быть ограничены сверху и снизу.

* **hp.loguniform**(label, low, high): Возвращает значение, распределенное согласно exp(uniform(low, high)), так что логарифм возвращаемого значения равномерно распределен.

* **hp.qloguniform**(label, low, high, q): Возвращает значение, похожее на round(exp(uniform(low, high)) / q) * q. Подходит для дискретных переменных, для которых целевая функция "гладкая" и становится гладкой с увеличением значения.

* **hp.normal**(label, mu, sigma): Возвращает вещественное значение, распределенное нормально со средним значением mu и стандартным отклонением sigma.

* **hp.qnormal**(label, mu, sigma, q): Возвращает значение вроде round(normal(mu, sigma) / q) * q. Подходит для дискретных переменных, которые вероятно принимают значение вокруг mu, но в основном неограничены.

* **hp.lognormal**(label, mu, sigma): Возвращает значение, распределенное согласно exp(normal(mu, sigma)), так что логарифм возвращаемого значения распределен нормально.

* **hp.qlognormal**(label, mu, sigma, q): Возвращает значение, похожее на round(exp(normal(mu, sigma)) / q) * q. Подходит для дискретных переменных, для которых целевая функция гладкая и становится гладкой с увеличением размера переменной.

```python
# Определяем пространство гиперпараметров
space = {
    'units': hp.choice('units', [64, 128, 256, 512]),
    'filters': hp.choice('filters', [32, 64, 128]),
    'kernel_size': hp.choice('kernel_size', [3, 5]),
    'pool_size': hp.choice('pool_size', [2, 3]),
    'dropout': hp.uniform('dropout', 0, 0.5),
    'batch_size': hp.choice('batch_size', [32, 64, 128]),
    'learning_rate': hp.loguniform('learning_rate', -5, -2),
    'activation': hp.choice('activation', ['relu', 'tanh'])
}
```

В настоящее время в Hyperopt реализованы три алгоритма:

* **Random Search** (Случайный поиск): Это самый простой алгоритм, который перебирает гиперпараметры случайным образом. Он не использует информацию из предыдущих прогонов для выбора следующего набора гиперпараметров. Из-за своей простоты и отсутствия какого-либо предположения о лучших гиперпараметрах, этот алгоритм может быть полезным в случаях, когда у нас нет знаний о лучших гиперпараметрах.

* **Tree of Parzen Estimators** (TPE, Дерево оценок Parzen): TPE - это более сложный алгоритм, который использует результаты предыдущих прогонов для выбора гиперпараметров для следующего прогона. Алгоритм TPE начинает с пробного значения гиперпараметра из пространства поиска и оценивает его с помощью целевой функции. Затем он обновляет модель вероятностной оценки на основе результатов испытания и выбирает новое пробное значение с учетом найденной модели.

Основная идея TPE заключается в поиске баланса между исследованием (exploration) и использованием (exploitation) пространства гиперпараметров. Он стремится исследовать области пространства, где находится больше успешных испытаний, а также использовать значения гиперпараметров, которые показывают хорошие результаты.

* **Adaptive TPE** (Адаптивный TPE): Адаптивный TPE - это модификация TPE, которая адаптируется к конкретной задаче для ускорения поиска.
```python
import numpy as np

def getBestModelfromTrials(trials):
    valid_trial_list = [trial for trial in trials
                            if STATUS_OK == trial['result']['status']]
    losses = [ float(trial['result']['loss']) for trial in valid_trial_list]
    index_having_minumum_loss = np.argmin(losses)
    best_trial_obj = valid_trial_list[index_having_minumum_loss]
    return best_trial_obj['result']['Trained_Model']
```
